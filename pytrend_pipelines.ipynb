```python
from pytrends.request import TrendReq
from pytrends import dailydata
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### - monthly breakdown

```python
# Define the date range for the current month (November 2023)
start_year = 2023
start_month = 10  # November
stop_year = 2023
stop_month = 11  # November

# Specify the keyword and geo location
keyword = "Prabowo Gibran"
geo = 'ID'  # Geo code for Indonesia

# Fetch daily data for the specified keyword and date range
df = dailydata.get_daily_data(keyword, start_year, start_month, stop_year, stop_month, geo=geo)

print(df)
```

```python
# Set the figure size
plt.figure(figsize=(12, 6))

# Plot the daily search volume data
sns.lineplot(data=df, x=df.index, y="Prabowo Gibran", label="Search Volume")

# Set labels and title
plt.xlabel("Date")
plt.ylabel("Search Volume")
plt.title("Daily Search Volume for 'Prabowo Gibran' in Indonesia (November 2023)")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.legend()
plt.tight_layout()
plt.show()
```
